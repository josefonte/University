```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df_powercap_1_1, df_powercap_1_12, etc. are DataFrames containing the data

# Calculate means for each program and power limit combination (Time in milliseconds)
spectral_time_means = [
    df_powercap_1_1['Time (ms)'].mean(),
    df_powercap_1_12['Time (ms)'].mean(),
    df_powercap_1_15['Time (ms)'].mean(),
    df_powercap_1_25['Time (ms)'].mean(),
    df_powercap_1_47['Time (ms)'].mean(),
    df_powercap_1_55['Time (ms)'].mean()
]

shootout_time_means = [
    df_powercap_2_1['Time (ms)'].mean(),
    df_powercap_2_12['Time (ms)'].mean(),
    df_powercap_2_15['Time (ms)'].mean(),
    df_powercap_2_25['Time (ms)'].mean(),
    df_powercap_2_47['Time (ms)'].mean(),
    df_powercap_2_55['Time (ms)'].mean()
]

imaginary_time_means = [
    df_powercap_3_1['Time (ms)'].mean(),
    df_powercap_3_12['Time (ms)'].mean(),
    df_powercap_3_15['Time (ms)'].mean(),
    df_powercap_3_25['Time (ms)'].mean(),
    df_powercap_3_47['Time (ms)'].mean(),
    df_powercap_3_55['Time (ms)'].mean()
]

real_time_means = [
    df_powercap_4_1['Time (ms)'].mean(),
    df_powercap_4_12['Time (ms)'].mean(),
    df_powercap_4_15['Time (ms)'].mean(),
    df_powercap_4_25['Time (ms)'].mean(),
    df_powercap_4_47['Time (ms)'].mean(),
    df_powercap_4_55['Time (ms)'].mean()
]

# Set up bar chart parameters
bar_width = 0.15
bar_positions = np.arange(6)

# Plot bar chart
plt.bar(bar_positions - bar_width, spectral_time_means, width=bar_width, label='Spectral')
plt.bar(bar_positions, shootout_time_means, width=bar_width, label='Shootout')
plt.bar(bar_positions + bar_width, imaginary_time_means, width=bar_width, label='Imaginary')
plt.bar(bar_positions + 2*bar_width, imaginary_time_means, width=bar_width, label='Real')

# Set labels and title
plt.ylabel('Mean Time (ms)')
plt.title('Execution Time per Program and Power Limit')
plt.legend(title='Program')
plt.xticks(bar_positions, ['Power Limit -1', 'Power Limit 12', 'Power Limit 15', 'Power Limit 25', 'Power Limit 47', 'Power Limit 55'], rotation=45)

# Show the plot
plt.show()

```
