```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 

columns = ['Language', 'Program', 'PowerLimit', 'Package', 'Core(s)', 'GPU', 'DRAM', 'Time (ms)', 'Temperature', 'Memory']

powercap_1 =  '1.csv'
powercap_12 = '2.csv'
powercap_15 = '5.csv'
powercap_25 = '10.csv'
powercap_47 = '25.csv'
powercap_55 = '50.csv'



# Read CSV file into a DataFrame
df_powercap_1_1 = pd.read_csv (powercap_1 , names=columns, skiprows=1)
df_powercap_1_12 = pd.read_csv(powercap_12, names=columns, skiprows=1)
df_powercap_1_15 = pd.read_csv(powercap_15, names=columns, skiprows=1)
df_powercap_1_25 = pd.read_csv(powercap_25, names=columns, skiprows=1)
df_powercap_1_47 = pd.read_csv(powercap_47, names=columns, skiprows=1)
df_powercap_1_55 = pd.read_csv(powercap_55, names=columns, skiprows=1)

```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df_powercap_1_1, df_powercap_1_12, etc. are DataFrames containing the data

# Calculate means for each program and power limit combination (Time in milliseconds)
spectral_time_means = [
    df_powercap_1_1['Package'].mean(),
    df_powercap_1_12['Package'].mean(),
    df_powercap_1_15['Package'].mean(),
    df_powercap_1_25['Package'].mean(),
    df_powercap_1_47['Package'].mean(),
    df_powercap_1_55['Package'].mean()
]


# Set up bar chart parameters
bar_width = 0.2
bar_positions = np.arange(6)

# Plot bar chart
plt.bar(bar_positions - bar_width, spectral_time_means, width=bar_width, label='Spectral')


# Set labels and title
plt.ylabel('Mean Value')
plt.title('Energy Consumption')
plt.legend(title='Program')
plt.xticks(bar_positions, ['Power Limit -1', 'Power Limit 2', 'Power Limit 5', 'Power Limit 10', 'Power Limit 25', 'Power Limit 50'], rotation=45)

# Show the plot
plt.show()

```
